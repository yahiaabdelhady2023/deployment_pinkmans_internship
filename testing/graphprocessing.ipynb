```python
import pandas as pd

data = pd.read_csv('DB Export 1.csv', parse_dates=['Order Date'], index_col='Order Date')
data=data[(data.Venue.str.contains('Whiteladies')) & (data['PLU Name']=='Pain Au Chocolat')]
data['Sold Items']=1
data=data.groupby(['Order Date'])[['Net Sale','Sold Items']].sum()
data
```

```python

# Aggregate by week
data = data.resample('W-SUN').agg({
    'Net Sale': 'sum',      # Sum Net Sales
    'Sold Items': 'sum'     # Sum Sold Items
})
data
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Sample structure of your dataset

df = pd.DataFrame(data)

# Convert 'Order Date' to datetime and set as index
# df['Order Date'] = pd.to_datetime(df['Order Date'])
df = df.sort_values(by='Order Date')

# Aggregate by 'Order Date' for total daily Net Sales
# daily_sales = df.groupby('Order Date')['Net Sale'].sum().reset_index()
# daily_sales.set_index('Order Date', inplace=True)

# Check for stationarity using ADF test
daily_sales=df
def check_stationarity(series):
    result = adfuller(series)
    print(f"ADF Statistic: {result[0]:.2f}")
    print(f"p-value: {result[1]:.4f}")
    if result[1] > 0.05:
        print("Series is non-stationary, consider differencing.")
    else:
        print("Series is stationary.")

check_stationarity(daily_sales['Net Sale'])

# Split into training and testing sets
train_size = int(len(daily_sales) * 0.8)  # 80% train, 20% test split
train, test = daily_sales.iloc[:train_size], daily_sales.iloc[train_size:]

# Define rolling forecast function
def rolling_forecast(train, test, order=(1, 1, 1)):
    history = train['Net Sale'].tolist()  # Initial training history
    predictions = []
    
    for t in range(len(test)):
        try:
            # Fit the ARIMA model
            model = ARIMA(history, order=order)
            model_fit = model.fit()
            
            # Make a forecast
            forecast = model_fit.forecast(steps=1)
            if isinstance(forecast, (np.ndarray, list)):  # Handle different return types
                forecast = forecast[0]
            predictions.append(forecast)
            
            # Append the actual observation to the history
            history.append(test['Net Sale'].iloc[t])
        except Exception as e:
            print(f"Error on step {t}: {e}")
            predictions.append(np.nan)  # Append NaN if an error occurs
            continue
    
    return predictions

# Rolling forecast with ARIMA
order = (1, 1, 4)  # ARIMA (p, d, q) parameters
predictions = rolling_forecast(train, test, order=order)

# Evaluate the model if predictions are valid
if not all(np.isnan(p) for p in predictions):  # Ensure predictions are not all NaN
    mse = mean_squared_error(test['Net Sale'], predictions)
    rmse = np.sqrt(mse)
    print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['Net Sale'], label='Actual Sales')
plt.plot(test.index, predictions, label='Predicted Sales', linestyle='--', color='orange')
plt.legend()
plt.title('Rolling Forecast for Net Sales')
plt.xlabel('Order Date')
plt.ylabel('Net Sales')
plt.grid()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import pickle

def update_and_predict_two_weeks(train_data, new_week_data, order=(1, 1, 1)):
    """
    Update ARIMA model with the last two weeks of data and predict the next week's sales.
    """
    # Append the new week data to the training data
    new_week_data['Order Date'] = pd.to_datetime(new_week_data['Order Date'])
    new_week_data.set_index('Order Date', inplace=True)
    train_data = train_data.append(new_week_data)

    # Only use the last two weeks of data for training
    last_two_weeks_data = train_data.tail(2)

    # Fit ARIMA model with the last two weeks of data
    model = ARIMA(last_two_weeks_data['Net Sale'], order=order)
    model_fit = model.fit()

    # Save the updated model
    pickle.dump(model_fit, open('arima_model_updated.pkl', 'wb'))

    # Forecast next week's sales (1 step ahead)
    forecast = model_fit.forecast(steps=1)
    return forecast[0]

# Example: Load historical data
train_data = pd.read_csv("historical_sales_data.csv", parse_dates=['Order Date'], index_col='Order Date')

# Example of new weekly data (this week)
new_week_data = pd.DataFrame({
    'Order Date': ['2023-09-22'],
    'Net Sale': [550.0],
    'Sold Items': [120]
})

# Predict next week's sales using the last two weeks of data
next_week_forecast = update_and_predict_two_weeks(train_data, new_week_data)
print(f"Predicted Net Sale for next week: {next_week_forecast}")
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

def train_and_predict_next_week(train_data, order=(1, 1, 1)):
    """
    Train the ARIMA model on the last 14 days of data and predict the next week's daily sales.
    """
    # Ensure the 'Order Date' column is datetime
    train_data['Order Date'] = pd.to_datetime(train_data['Order Date'])
    train_data.set_index('Order Date', inplace=True)
    
    # Use only the last 14 days of data for training
    last_two_weeks_data = train_data.tail(14)
    
    # Train the ARIMA model
    model = ARIMA(last_two_weeks_data['Net Sale'], order=order)
    model_fit = model.fit()

    # Forecast the next 7 days
    forecast = model_fit.forecast(steps=7)
    
    # Generate dates for the next 7 days
    forecast_dates = pd.date_range(last_two_weeks_data.index[-1] + pd.Timedelta(days=1), periods=7)

    # Prepare the forecast as a DataFrame
    forecast_df = pd.DataFrame({
        'Date': forecast_dates,
        'Predicted Net Sale': forecast
    })

    return forecast_df

# Example: Load historical data (replace with your own dataset)
train_data = pd.read_csv("historical_sales_data.csv", parse_dates=['Order Date'], index_col='Order Date')

# Call the function to predict the next week's daily sales
next_week_forecast = train_and_predict_next_week(train_data)

# Display the forecast
print(next_week_forecast)

# Optionally, plot the predictions
next_week_forecast.plot(x='Date', y='Predicted Net Sale', kind='line', title="Next Week's Daily Sales Forecast", xlabel="Date", ylabel="Predicted Net Sale")
plt.show()
```
